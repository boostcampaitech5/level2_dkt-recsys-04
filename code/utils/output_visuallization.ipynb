```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CSV 파일 경로 설정
csv_file = "/opt/ml/level2_dkt-recsys-04/code/boosting/outputs/CatBoost/output.csv"

# CSV 파일 읽기
data = pd.read_csv(csv_file)

# 막대 그래프 그리기
plt.bar(data['id'], np.where(data['prediction'] >= 0.5, 1, 0))#np.where(pred >= 0.5, 1, 0)

# 그래프 타이틀 및 축 제목 설정
plt.title("Output CatBoost")
plt.xlabel("userID")
plt.ylabel("prediction")

# 그래프 출력
plt.show()
```

```python
# 막대 그래프 그리기
plt.bar(data['id'], data['prediction'])#np.where(pred >= 0.5, 1, 0)

# 그래프 타이틀 및 축 제목 설정
plt.title("Output CatBoost")
plt.xlabel("userID")
plt.ylabel("prediction")

# 그래프 출력
plt.show()
```

```python
# displot으로 분포 시각화
sns.displot(data['prediction'])

# 그래프 출력
sns.plt.show()
```
